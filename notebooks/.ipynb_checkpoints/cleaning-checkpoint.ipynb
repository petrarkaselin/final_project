```python
def convert_to_snake_case(input_string):
    # Add an underscore before any uppercase letter and convert to lowercase
    snake_case_string = re.sub(r'([A-Z])', r'_\1', input_string).lower()

    # Remove leading underscore if present
    snake_case_string = snake_case_string.lstrip('_')

    return snake_case_string
```

```python
# Apply the function to all column names
df.columns = [convert_to_snake_case(col) for col in df.columns]
```

```python
df.head()
```

```python
# Display columns with fewer than 10 unique values
for column in df.columns:
    unique_values = df[column].unique()
    if len(unique_values) < 10:
        print(f"Unique values in '{column}': {unique_values}")
```

```python
# Drop columns with only one unique value
columns_to_drop = [col for col in df.columns if df[col].nunique() == 1]
df = df.drop(columns=columns_to_drop)
```
